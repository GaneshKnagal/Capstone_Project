3.8.2	Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.

```python
# SQL query to find the district in each State/UT with the highest number of murders year-wise
sql_query = """
    WITH RankedMurders AS (
        SELECT
            "STATE/UT" AS STATE_UT,
            YEAR,
            DISTRICT,
            MURDER,
            ROW_NUMBER() OVER (PARTITION BY "STATE/UT", YEAR ORDER BY MURDER DESC) AS RankByMurder
        FROM
            {}
    )
    SELECT
        STATE_UT,
        YEAR,
        DISTRICT,
        MURDER
    FROM
        RankedMurders
    WHERE
        RankByMurder = 1
    ORDER BY
        STATE_UT, YEAR;
""".format(new_table_name)

# Execute the query and fetch the result into a DataFrame
result_df = pd.read_sql(sql_query, engine)

# Display the result
result_df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Assuming df is your DataFrame with crime-related columns
crime_features = ['Murder_x', 'Rape_x','Kidnapping and Abduction_x','Procuration of minor girls','Buying of girls for prostitution', 'Selling of girls for prostitution','Other Crimes', 'Total', 'Rape_y',
       'Kidnapping and Abduction_y','Assault on women with intent to outrage her modesty','Insult to modesty of Women','MURDER', 'ATTEMPT TO MURDER',
       'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 'RAPE', 'CUSTODIAL RAPE','OTHER RAPE', 'KIDNAPPING & ABDUCTION',
       'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS','ROBBERY', 'BURGLARY', 'THEFT','OTHER THEFT', 'RIOTS','TOTAL IPC CRIMES',
       'Murder_y', 'Rape_x', 'Kidnapping and Abduction','Other Crimes Against SCs','Murder', 'Rape_y', 'Kidnapping Abduction', 'Dacoity_y', 'Robbery_y',
       'Arson_y', 'Hurt_y','Prevention of atrocities (POA) Act_y', 'Other Crimes Against STs']

# Select relevant features
df_features = df[crime_features]

# Impute missing values
imputer = SimpleImputer(strategy='mean')
df_features_imputed = imputer.fit_transform(df_features)

# Feature Scaling
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features_imputed)

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Analyze the clusters and assign labels
cluster_labels = {
    0: "Sensitive Area's",
    1: "Peaceful Area's",
    2: "Moderate Area's"
}

df['Cluster_Label'] = df['Cluster'].map(cluster_labels)

# Display the resulting DataFrame with cluster assignments
df[['STATE/UT', 'DISTRICT', 'Cluster_Label']]
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Plot the clusters using the first two features
plt.figure(figsize=(12, 8))
sns.scatterplot(x=df_scaled[:, 0], y=df_scaled[:, 1], hue=df['Cluster_Label'], palette='viridis', s=100)

# Plot the cluster centers
centers = kmeans.cluster_centers_
sns.scatterplot(x=centers[:, 0], y=centers[:, 1], marker='X', s=200, color='red', label='Cluster Centers')

# Add labels and title
plt.title('K-Means Clustering of Districts based on Crime Features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
